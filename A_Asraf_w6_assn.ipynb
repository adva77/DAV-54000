###   6: Make a bar plot of the â€˜cutâ€™ data. What does it tell us about the frequencies of the individual cut values within the data set?

***

I used numpy in order to create the bar chart. I set my columns to the different cut types, and frequencies represent how many of that specific cut was represented in the CSV file 
According to the below bar chart, one can see that most diamons are ideal, followed by premium. The least common cut of diamond was "fair". One can assume that when a buyer buys a diamond, he preferes to spend his money wisely and to buy a good quality diamond. 

```python
import numpy as np
frequency = data_frame['cut'].value_counts()
objects = ['Ideal', 'Premium', 'Very Good', 'Good', 'Fair']
y_pos = np.arange(len(objects))
new=[]
for i in range(5):
    new.append(frequency[i])

plt.bar(y_pos, new, align='center', alpha=0.5)
plt.xticks(y_pos,objects)
plt.xlabel('Cut')
plt.ylabel('Frequencies of cut values') 
plt.show()
```
