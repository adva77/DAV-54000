### Tidying and Transforming data:

I have created a table that orders the students and marks according to the different subjects so it will be easier to analyze and derive insights from. Also, it is a neater form of presenting the data. 

```python
data = pd.melt(data_1, ['Student'])
data=data.rename(columns={'variable':'Subjects','value':'Marks'})
data
```

### Analysis:

1. Average score of each student.
2. Minimum and maximum scores in each subject.

```python
d=data.groupby(['Student','Subjects'])[['Marks']].sum()
d
```

```python
new = d.groupby(['Student']).agg('mean')
new.columns = ['Average scores']
new
```

```python
new2 = d.groupby(['Subjects']).agg(['min','max'])
new2
```

```python
data_3
```

### Tidying and Transforming the data:
I have created a table in which the different test scores are ordered by the test number. 

```python
data = pd.melt(data_3,['Name'])
data = data.rename(columns = {'variable':'Tests','value':'Result'})
data
```

# Analysis:

1. Find student with maximum average score.
2. Which student has the greatest range of tests?
3. Which test has the highest grades?

```python
new = data.groupby(['Name','Tests'])[['Result']].sum()
new
```

```python
avg = new.groupby(['Name']).agg('mean')
avg
```

```python
marks=avg.loc[avg['Result'].idxmax()]
name = avg.idxmax()
print(name[0], 'has the maximum average score i.e.,',marks[0])
```

```python
count = data.groupby(['Name']).count()
count = count.rename(columns = {'Result':'Tests Taken by each student'})
count = count.drop(['Tests'],axis=1)
count
```

```python
count = count[count['Tests Taken by each student'] >= 4]
print('Following are the students with maximum range of tests\n')
print(count)
```

```python
maxx = new.groupby(['Tests']).sum()
print(maxx)
max_grade = maxx.idxmax()
print('\n',max_grade[0], 'has maximum grades.')
```

### Tidying and Transforming:

I have ordered the data by ethnicity since the example above was a little messy and harder to read and analyze. 

```python
index = ['Unique ID of the institution','Level of student']
data = pd.melt(data_2,index,var_name = 'Ethnicity',value_name='count')
data
```

```python
new = data.groupby(['Unique ID of the institution','Level of student','Ethnicity']).sum()
new
```

# Analysis

1. Overall maximum number of enrollments with respect to ethnicity.
2. Maximum enrollments with respect to the level of student and ethnicity.

```python
maxx = new.groupby(['Ethnicity']).sum()
maxx = maxx.rename(columns = {'count':'Total enrollments'})
maxx
```

```python
name = maxx.idxmax()
count = maxx.max()
print(name[0], 'has the maximum number of enrollments i.e.,',count[0])
```

```python
new2 = new.groupby(['Level of student','Ethnicity']).sum()
new2
```

```python
level_max = new2.groupby(['Level of student']).max()
level_max = level_max.rename(columns = {'count':'Maximum enrollments'})

new2 = new2.groupby(['Level of student']).idxmax()
new2 = new2.rename(columns = {'count':'(level, ethnicity)'})
new2
```
